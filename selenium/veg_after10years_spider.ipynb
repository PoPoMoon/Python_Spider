```python
from selenium import webdriver 
from selenium.webdriver.common.by import By
from time import sleep
import selenium.webdriver.support.ui as ui
import os

options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")       
options.add_argument("--incognito")
options.add_argument("--disable-popup-blocking ")

# 下面為更改下載目錄位址
out_path="D:\\Download\\Farm\\Cabbage"    # 後面可以改目前要放在哪個目錄(如果沒有目錄會自己建立)


prefs = {'profile.default_content_settings.popups': 0, 'download.default_directory': out_path}
options.add_experimental_option('prefs', prefs)
driver = webdriver.Chrome(options = options)

url='http://amis.afa.gov.tw/veg/VegProdDayTransInfo.aspx'
driver.get(url)


def main():
    produtc_code='LA1'                  # 產品代號
    vegetable_name='Cabbage'            # 下載檔案改名成 100yearXXXX 
    
    go_web=get_after_100()
    go_web.market()
    go_web.product(produtc_code)
    for i in range(16):
        print('目前在第',go_web.now_year,'年,count=',go_web.now_year_count)
        go_web.start_year()
        go_web.end_year()
        go_web.downlod()
        go_web.change_name(vegetable_name)
        go_web.subtract()
        
    driver.close()
        
class get_after_100():

    def __init__(self):
        self.now_year=100
        self.now_year_count=15
    
    def subtract(self):
        self.now_year-=1
        self.now_year_count-=1
    
    def start_year(self):
        sleep(1)
        driver.find_element(By.CSS_SELECTOR,"input#ctl00_contentPlaceHolder_ucDateScope_rblDateScope_1").click()
        sleep(1)
        driver.find_element(By.CSS_SELECTOR,"input#ctl00_contentPlaceHolder_txtSTransDate").click()
        self.the_same_click()
        sleep(1)
        driver.find_elements(By.CSS_SELECTOR,"select#sMonth_amis option")[0].click()
        sleep(1)
        start_jan_day=driver.find_elements(By.CSS_SELECTOR,"div#calDiv_amis tr td font[face='Arial Black']")
        count=0
        for i in start_jan_day:
            if i.text=='1':
                break
            count+=1
        start_jan_day[count].click()

    def end_year(self):
        sleep(1)
        driver.find_element(By.CSS_SELECTOR,"input#ctl00_contentPlaceHolder_txtETransDate").click()
        self.the_same_click()
        sleep(1)
        driver.find_elements(By.CSS_SELECTOR,"select#sMonth_amis option")[-1].click()
        sleep(1)
        end_jan_day=driver.find_elements(By.CSS_SELECTOR,"div#calDiv_amis tr td font[face='Arial Black']")
        count=34
        for i in end_jan_day[::-1]:
            if i.text=='31':
                break
            count-=1
        end_jan_day[count].click()

    def the_same_click(self):
        sleep(1)
        driver.find_element(By.CSS_SELECTOR,"select#sYear_amis").click()
        sleep(1)
        driver.find_elements(By.CSS_SELECTOR,"select#sYear_amis option")[self.now_year_count].click()
        sleep(1)
        driver.find_element(By.CSS_SELECTOR,"select#sMonth_amis").click()
        
    
    def market(self):
        sleep(1)
        driver.find_element(By.CSS_SELECTOR,"textarea#ctl00_contentPlaceHolder_txtMarket").click()
        sleep(1)

#       這裡為切換frame
        iframe = driver.find_element_by_css_selector('iframe')
        driver.switch_to_frame(iframe)
    
        driver.find_element(By.CSS_SELECTOR,"input#radlMarketRange_0").click()
        
    def product(self,code):
        sleep(1)
        driver.find_element(By.CSS_SELECTOR,"textarea#ctl00_contentPlaceHolder_txtProduct").click()
        sleep(1)
    
#       這裡為切換frame
        iframe = driver.find_element_by_css_selector('iframe')
        driver.switch_to_frame(iframe)
        
        textInput = driver.find_element(By.CSS_SELECTOR,"html input#txtKeyWord")  # 這裡會出現warning
        textInput.send_keys(code)
        sleep(1)
        
        driver.find_element(By.CSS_SELECTOR,"input#btnQuery").click()
        sleep(1)
        driver.find_element(By.CSS_SELECTOR,"select#lstProduct option").click()
        sleep(1)
        driver.find_element(By.CSS_SELECTOR,"input#btnConfirm").click()        
        
    def downlod(self):
        sleep(1)
        driver.find_element(By.CSS_SELECTOR,"input#ctl00_contentPlaceHolder_btnXls").click()
        print('sleep 3 秒')
        sleep(3)
    
    def change_name(self,vege_name):
        for fname in os.listdir(out_path):
            if fname=='蔬菜產品日交易行情.xls':
                new_fname = f'{self.now_year}year{vege_name}'        # 要改什麼名字在這改
                os.rename(os.path.join(out_path, fname), os.path.join(out_path, new_fname))
            
            
if __name__=='__main__':
    main()
```
